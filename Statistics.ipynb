#### Conditional Probability refers to the probability of event A, given information about the occurrence of another event B

In this case, we need to find the conditional probability of people who planned to purchase (Yes) and actually placed an order (Yes) given that they planned to purchase.

#### Probability of A given B is written as P(A | B).

Probability of A given B is written as P(A | B).

$P(A\mid B) = \frac{P(A \quad and \quad B)}{P(B)}$

where P(A and B) = Joint probability of A and B
*     P(A)       = Marginal probability of A
*     P(B)       = Marginal probability of B

```python
# P1 = P(Actually placed an order | Planned to purchase)
# P2 = Planned to purchase and actually placed an order
# P3 = Planned to purchase
# Number of people who planned to purchase and placed an order (Yes) / Number of people who planned to purchase (Yes) 
a1   = (400 / 2000 )/ (500 / 2000)

print('P(Actually placed an order | Planned to purchase) is %1.4f' % a1)  
```

## Q2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions. (4 points)
1. Probability that none of the items are defective? (1 point)
2. Probability that exactly one of the items is defective? (1 point)
3. Probability that two or fewer of the items are defective? (1 point)
4. Probability that three or more of the items are defective ? (1 point)


```python
import numpy as np
import pandas  as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.stats import binom
p   =  0.05 # failure rate of products that the company manufactures is 5%
n   =  10 # sample size
k   =  np.arange(0,11) # An array of quantiles

binomial = stats.binom.pmf(k,n,p)
print(binomial)
```

```python
# None of the manufactured items are defective
bp0 = binomial[0]
percentage = bp0 * 100
print(f"Binomial Probability P(X = 0): {bp0:.6f}")
print(f"Percentage: {percentage:.2f}%")
```

```python
# Exactly one of the manufactured itemss defective?
bp1 = binomial[1]
percentage = bp1 * 100
print(f"Binomial Probability P(X = 1): {bp1:.6f}")
print(f"Percentage: {percentage:.2f}%")
```

```python
# Two or fewer of the manufactured items are defective
# Calculate the cumulative binomial probability P(X ≤ k)
cumbinomial = stats.binom.cdf(k,n,p)
D = cumbinomial[2]
percentage = D * 100
print(f"Binomial Probability P(X = 1): {D:.6f}")
print(f"Percentage: {percentage:.2f}%")
```

```python
# Three or more of the manufactured items are defective
# Calculate the complementary cumulative binomial probability P(X ≥ k+1)
P = 1- cumbinomial[2]
percentage = P * 100

print(f"Complementary Cumulative Probability (P(X ≥ {k+1})): {P:.6f}")
print(f"Percentage: {percentage:.2f}%")
```

## Q4. Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed, marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures the percentage of orders that are taken correctly is 86.8%.  Suppose that you place an order with the bot and two friends of your independently place orders with the same bot. Answer the following questions.

1. What is the probability that all three orders will be recognised correctly.

2. What is the probability that none of the three orders will be recognised correctly.

3. What is the probability that at least two of the three orders will be recognised correctly?

Given that the recognition accuracy is 86.8%, p = 0.868 represents the chance of a correct recognition, whereas q = 1 - p = 0.132 represents the likelihood of an incorrect recognition.


* a. The likelihood that all three orders will be accurately identified:
Since there were independent events in this scenario, we just needed to multiply the probabilities:
P(all right) equals `p ** 3`



* b. Chance that none of the three orders will be accurately identified:
In the same way, for three separate wrong recognitions:
P(none accurate) equals `q ** 3`


* c. The likelihood that at least two of the three orders will be accurately identified:
There are three possible outcomes for this: all three correct, all three incorrect, or two correct and one incorrect. Add up the calculations for each scenario:
P(at least two accurate) is equal to `p**3 + 3*p**2*q`

```python
# Given recognition accuracy
p = 0.868  # Probability of correct recognition
q = 1 - p  # Probability of incorrect recognition

# A. Probability that all three orders will be recognized correctly
prob_all_correct = p ** 3

# B. Probability that none of the three orders will be recognized correctly
prob_none_correct = q ** 3

# C. Probability that at least two of the three orders will be recognized correctly
prob_at_least_two_correct = p**3 + 3*p**2*q

# Convert probabilities to percentages
prob_all_correct_percentage = prob_all_correct * 100
prob_none_correct_percentage = prob_none_correct * 100
prob_at_least_two_correct_percentage = prob_at_least_two_correct * 100

# Printing the results
print("A. Probability that all three orders will be recognized correctly: %.2f%%" % prob_all_correct_percentage)
print("B. Probability that none of the three orders will be recognized correctly: %.2f%%" % prob_none_correct_percentage)
print("C. Probability that at least two of the three orders will be recognized correctly: %.2f%%" % prob_at_least_two_correct_percentage)
```

## Q5. A group of 300 professionals sat for a competitive exam. The results show the information of marks obtained by them have a mean of 60 and a standard deviation of 12. The pattern of marks follows a normal distribution. Answer the following questions. (3 points)
1. What is the percentage of students who score more than 80? (1 point)
2. What is the percentage of students who score less than 50? (1 point)
3. What should be the distinction mark if the highest 10% of students are to be awarded distinction? (1 point)

```python
# Given parameters
mean = 60
std_dev = 12

# Calculate the z-score for a given value (x) using the formula: z = (x - mean) / std_dev
def calculate_z_score(x):
    return (x - mean) / std_dev

# 1. Percentage of students who score more than 80
x_more_than_80 = 80
z_more_than_80 = calculate_z_score(x_more_than_80)
prob_more_than_80 = 1 - stats.norm.cdf(z_more_than_80)

# 2. Percentage of students who score less than 50
x_less_than_50 = 50
z_less_than_50 = calculate_z_score(x_less_than_50)
prob_less_than_50 = stats.norm.cdf(z_less_than_50)

# 3. Distinction mark for the highest 10% of students
p_distinction = 0.1
distinction_mark = stats.norm.ppf(1 - p_distinction) * std_dev + mean

# Print the results
print("1. Percentage of students who score more than 80: %.2f%%" % (prob_more_than_80 * 100))
print("2. Percentage of students who score less than 50: %.2f%%" % (prob_less_than_50 * 100))
print("3. Distinction mark for the highest 10%% of students: %.2f" % distinction_mark)
```

### Univariate Analysis

```python
def histogram_boxplot(data, feature, figsize=(15, 10), kde=False, bins=None):
    """
    Boxplot and histogram combined

    data: dataframe
    feature: dataframe column
    figsize: size of figure (default (15,10))
    kde: whether to show the density curve (default False)
    bins: number of bins for histogram (default None)
    """
    f, (ax_box, ax_hist) = plt.subplots(
        nrows=2,  # Number of rows of the subplot grid = 2
        sharex=True,  # x-axis will be shared among all subplots
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize,
    )  # creating the 2 subplots
    
    sns.boxplot(
        data=data, x=feature, ax=ax_box, showmeans=True, color="violet"
    )  # boxplot will be created and a triangle will indicate the mean value of the column
    
    if bins:
        sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist, bins=bins)
    else:
        sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist)  # For histogram
    
    ax_hist.axvline(data[feature].mean(), color="green", linestyle="--")  # Add mean to the histogram
    ax_hist.axvline(data[feature].median(), color="black", linestyle="-")  # Add median to the histogram
```
